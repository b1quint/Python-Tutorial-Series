## MatPlotLib

### A simple plot

Now that we know that we can do some maths easily using `numpy`, let us 
do some simple plots.

````python
from matplotlib.pyplot import *

plot(x_array, y_array)
show()
````

But we are **strongly** recommended to avoid the syntax above. The `from ... import *` goes against the Zen of Python since the imported functions are not explicit. You don't know what you are bringing and you might lose data. **Avoid this syntax at any cost!!**

An option commonly used is to use the `from ... import ... as ...` syntax. This way you keep the relationship between functions, variables, classes explicitly. Check below:

````python
from matplotlib import pyplot as plt

plt.plot(x_array, y_array)
plt.show()
````

### Basic Customization

You can easily customize most of the features in a MatPlotLib plot. 

````python
plt.plot(x_array, y_array, label='blue line')
plt.plot(x_array, y_array, 'C3o', label='circles')

plt.grid(alpha=0.25)
plt.legend(loc='best')

plt.xlabel('X')
plt.ylabel('Y')

_ = plt.title('My Plot')
````

### Customizing default parameters 

You can also change the plots behaviours using `plt.rcParams.update`:

````python
plt.rcParams.update({'font.size': 22})

plt.plot(x_array, y_array, label='blue line')
plt.plot(x_array, y_array, 'o', label='circles')

plt.grid(alpha=0.25)
plt.legend(loc='best')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('My Plot')
plt.show()

plt.rcParams.update({'font.size': 14})
````

### Using styles

MatPlotLib allows changing the style of the plots by simply using the `ply.style.context` **context manager**. Here is an example:

````python
with plt.style.context('seaborn'):

    plt.figure(dpi=300)
    plt.plot(x_array, y_array, label='blue line')
    plt.plot(x_array, y_array, 'o', label='circles')

    plt.grid()
    plt.legend(loc='best')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('My Plot')

    plt.show()
````

Or, alternatively, you can use the `plt.style.use` function but that will change the style of all future plots.

````python
plt.style.use('ggplot')

plt.plot(x_array, y_array, label='blue line')
plt.plot(x_array, y_array, 'o', label='circles')

plt.grid()
plt.legend(loc='best')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('My Plot')

plt.show()
````

### Plot Types

MatPlotLib has all sorts of graphics available. Here are some examples:

#### Scatter Plot

````python
np.random.seed(0)

x1 = np.random.normal(loc=0.5, scale=0.1, size=100)
y1 = np.random.normal(loc=0.5, scale=0.1, size=100)
z1 = np.random.random(100) * 200

x2 = np.random.normal(loc=0.75, scale=0.1, size=100)
y2 = np.random.normal(loc=0.75, scale=0.1, size=100)
z2 = np.random.random(size=100) * 200

with plt.style.context('seaborn-talk'):

    fig, ax = plt.subplots(dpi=130, figsize=(3, 3))

    ax.scatter(x1, y1, s=z1, alpha=0.5)
    ax.scatter(x2, y2, s=z2, alpha=0.5)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Scatter Plot')

    plt.show()
````

#### Histograms

````python
np.random.seed(0)

x1 = np.random.normal(loc=0.5, scale=0.1, size=100)
x2 = np.random.normal(loc=0.75, scale=0.15, size=100)

with plt.style.context('seaborn-talk'):
    
    fig, ax = plt.subplots(dpi=130, figsize=(3, 3))

    # Create the first histogram and save the bins for later
    n, bins, patches = ax.hist(x1, alpha=0.5, ec='C0', bins=20, 
                               range=[0, 1.5])

    # Create the second histogram using the bins from the first one
    ax.hist(x2, alpha=0.5, ec='C1', bins=bins)

    ax.set_xlabel('x')
    ax.set_title('Histogram')

    plt.show()
````

#### Error bars

````python
np.random.seed(0)

x = np.arange(-4, 5)
y = x ** 2 + 5 * x - 6 
error = np.random.random(y.size) * 10

p = np.polyfit(x, y, 3)
x_fit = np.linspace(x.min(), x.max(), 100)
y_fit = np.polyval(p, x_fit)

with plt.style.context('seaborn-talk'):
    
    fig, ax = plt.subplots(dpi=130, figsize=(3,3))
    
    ax.errorbar(x, y, marker='o', linestyle='none', xerr=0.5, yerr=error)
    ax.plot(x_fit, y_fit, label='Polyfit')
    ax.set_title('Errorbar')
    ax.legend(loc='upper left')
    
    plt.show()
````

## Wrapping Up

Let us now wrap everything together and use NumPy and MatPlotLib to create a plot that will contain the flux of sources in a stacked image versus the median value of the fluxes of the individual frames. This figure shall use a `scatter` plot with a color that will change depending on the signal to noise calculated (defined in some unknown fashion). 

Then, we will create a second figure that will be used to perform the same analysis but the Y axis will contain the ratio between the fluxes from the stacked frame and the median flux of each source in the individual frames.

### Reading the data

The data was reduced and stored in a text file. This file can be loaded using the numpy function `np.loadtxt`. We can check the beginning of the file by using the Unix command `head` invocked using the bang (`!`) character:

````python
!head data/my_table.txt
````

We don't need to load the whole data. For the first plot, we can do it only using the `FLUX_STACK`, the `MEDIAN` and the `SNR` columns. They are the columns number 4, 10 and 11, repectively (0-indexed).

````python
filename = 'data/my_table.txt'
flux_stack, flux_median, snr = np.loadtxt(
    filename, usecols=[4, 10, 11], unpack=True)
````

### First Plot

Now we want to have a scatter plot of `flux_stack` versus `flux_median`. Let's do this using Figures and Axes. It will be useful.

````python
fig, ax = plt.subplots(dpi=130)

ax.plot(flux_median, flux_stack)

plt.show()
````

### Changing Marker and LineStyle

Since the data is not necesarily ordered, we might want to change the line to use a round marker instead:

````python
fig, ax = plt.subplots(dpi=130)

ax.plot(flux_median, flux_stack, 'o')

plt.show()
````

Let's change the X and Y axis to a log scale for better visualization:

````python
fig, ax = plt.subplots(dpi=130)

ax.plot(flux_median, flux_stack, 'o')

ax.set_xscale('log')
ax.set_yscale('log')

plt.show()
````

### Adding Labels and Title

Much clearer! Now let's add a title, a grid and set the labels of both X and Y axis:

````python
fig, ax = plt.subplots(dpi=130)

ax.plot(flux_median, flux_stack, 'o')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel('Median of Sources on Individual Frames')
ax.set_ylabel('Flux of Sources in the Stacked Frame')
ax.set_title('Stacked Sources vs Median Fluxes')
ax.grid(alpha=0.5)

plt.show()
````

### Linear Fitting

The plots above confirm that the relationship between X and Y should be linear. We can check that by using `np.polyfit` and plotting the result:

````python
fig, ax = plt.subplots(dpi=130)

p = np.polyfit(flux_median, flux_stack, 1)

x_fit = np.linspace(flux_median.min(), flux_median.max(), 1000)
y_fit = np.polyval(p, x_fit)

ax.loglog(flux_median, flux_stack, 'o')

ax.plot(x_fit, y_fit, '-', 
        label=u'$f(x) = {0:.2f}\cdot x + {1:.2f}$'.format(*p))

ax.legend(loc='upper left')
ax.set_xlabel('Median of Sources on Individual Frames')
ax.set_ylabel('Flux of Sources in the Stacked Frame')
ax.set_title('Stacked Sources vs Median Fluxes')
ax.grid(alpha=0.5)

plt.show()
````

### Plotting ratio

Another way to see the result above is plotting $\frac{Flux Stack}{Flux Median} \times Flux Median$ and the flux should be around 1: 

````python
fig, ax = plt.subplots(dpi=130)

x = flux_median
y = flux_stack / flux_median

ax.plot(x, y, 'o')
ax.set_xscale('log')

ax.set_xlabel('Median of Sources on Individual Frames')
ax.set_ylabel('Stack Flux divided by Median Flux')

ax.set_title('Ratio between Stacked Sources and Median Fluxes vs Median Fluxes')

ax.grid(alpha=0.5)

plt.show()
````

### Adding a Violin Plot

It might be interesting to see the distribution of the points in the Y direction. For that, let us recreate our figure using `plt.subplots` with two columns that will share the y-axis. 

````python
x = flux_median
y = flux_stack / flux_median

fig, axes = plt.subplots(ncols=2, sharey=True, dpi=130)

ax1 = axes[1]
ax1.plot(x, y, marker='o', linestyle='none', alpha=0.4)
    
ax1.set_xscale('log')
ax1.set_xlabel('Flux median')
ax1.grid(alpha=0.25)

ax2 = axes[0]
ax2.violinplot(y, showmedians=True)
ax2.set_ylabel('Flux Stack / Flux Median')
ax2.set_xticks([])

plt.show()
````

This is pretty nice but we have to scale the plots so we can see the right axes better. This can be done passing the `gridspec_kw` argument, a dictionary used to control the axes grid. 

````python
x = flux_median
y = flux_stack / flux_median

fig, axes = plt.subplots(ncols=2, sharey=True, 
                         gridspec_kw={'width_ratios': [1, 15]}, dpi=130)

ax1 = axes[1]
ax1.plot(x, y, marker='o', linestyle='none', alpha=0.4)
    
ax1.set_xscale('log')
ax1.set_xlabel('Flux median')
ax1.grid(alpha=0.25)

ax2 = axes[0]
ax2.violinplot(y, showmedians=True)
ax2.set_ylabel('Flux Stack / Flux Median')
ax2.set_xticks([])

fig.suptitle('Stack Flux / Median Flux vs. Median Flux')
plt.show()
````

We can make the two plots closer to improve association between them using `fig.tight_layout`:

````python
x = flux_median
y = flux_stack / flux_median

fig, axes = plt.subplots(ncols=2, sharey=True, 
                         gridspec_kw={'width_ratios': [1, 15]}, dpi=130)

ax1 = axes[1]
ax1.plot(x, y, marker='o', linestyle='none', alpha=0.4)
    
ax1.set_xscale('log')
ax1.set_xlabel('Flux median')
ax1.grid(alpha=0.25)

ax2 = axes[0]
ax2.violinplot(y, showmedians=True)
ax2.set_ylabel('Flux Stack / Flux Median')
ax2.set_xticks([])
ax2.grid(alpha=0.25)

fig.suptitle('Stack Flux / Median Flux vs. Median Flux')
fig.tight_layout(w_pad=0)
plt.show()
````

Now the title is messed up. Let's fix that passing `rect` to the `fig.tight_layout` function:

````python
x = flux_median
y = flux_stack / flux_median

fig, axes = plt.subplots(ncols=2, sharey=True, 
                         gridspec_kw={'width_ratios': [1, 15]}, dpi=130)

ax1 = axes[1]
ax1.plot(x, y, marker='o', linestyle='none', alpha=0.4)
    
ax1.set_xscale('log')
ax1.set_xlabel('Flux median')
ax1.grid(alpha=0.25)

ax2 = axes[0]
ax2.violinplot(y, showmedians=True)
ax2.set_ylabel('Flux Stack / Flux Median')
ax2.set_xticks([])
ax2.grid(alpha=0.25)

fig.suptitle('Stack Flux / Median Flux vs. Median Flux')
fig.tight_layout(rect=[0, 0.05, 0.85, 0.95], w_pad=0)
plt.show()
````

### Using Scatter plot 

#### with different marker sizes

We can have more information by using `plt.scatter` instead of using `plt.plot`. `plt.scatter` allows passing an array to set the marker size parameter `s`:

````python
x = flux_median
y = flux_stack / flux_median
z = snr

fig, axes = plt.subplots(ncols=2, sharey=True, 
                         gridspec_kw={'width_ratios': [1, 15]}, dpi=140)

ax1 = axes[1]
ax1.scatter(x, y, s=z, marker='o', alpha=0.4)
    
ax1.set_xscale('log')
ax1.set_xlabel('Flux median')
ax1.grid(alpha=0.25)

ax2 = axes[0]
ax2.violinplot(y, showmedians=True)
ax2.set_ylabel('Flux Stack / Flux Median')
ax2.set_xticks([])
ax2.grid(alpha=0.25)

fig.suptitle('Stack Flux / Median Flux vs. Median Flux')
fig.tight_layout(rect=[0, 0.05, 0.85, 0.95], w_pad=0)
plt.show()
````

#### with different colors

Another option is to use colors instead of using marker size. This can be done passing the array to the `c` parameter instead:
